```python
import tensorflow as tf 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
plt.figure()
for i in range(100):
    plt.subplot(10, 10, i +1)
    plt.imshow(x_train[i])
    
# plt.imshow(x_train[0])
plt.show()
```

```python
# Data preprocessing 
# Flatten the data
x_train = np.reshape(x_train, [x_train.shape[0], -1])
x_test = np.reshape(x_test, [x_test.shape[0], -1])

# Convert labels to onehot vector
def one_hot(y, nb_class):
    temp = [0]*nb_class
    temp[y] = 1
    return temp

y_train_oh = np.array([one_hot(y, 10) for y in y_train], dtype=float)
y_test_oh = np.array([one_hot(y, 10) for y in y_test], dtype=float)

print(y_test[1000])
print(y_test_oh[1000])
print(y_train_oh.shape)
print(y_test_oh.shape)
```

```python
# 1. Build a computational graph
x_pl = tf.placeholder(tf.float32, [None, 784])
y_pl = tf.placeholder(tf.float32, [None, 10])

def fc_layer(x,input_shape, output_shape, name='fc',reuse =False):
    with tf.variable_scope(name):
        weight = tf.get_variable('weight', [input_shape, output_shape])
        bias = tf.get_variable('bias', [output_shape])
        output = tf.matmul(x, weight) + tf.reshape(bias, [1, output_shape])
        
        return output
    
h1=tf.nn.relu(fc_layer(x_pl, 784, 100, name='fc1'))
h2=tf.nn.relu(fc_layer(h1,100,100,name='fc2'))
predict = tf.nn.softmax(fc_layer(h2,100,10), name='fc3')
likelihood = tf.reduce_sum(predict*y_pl, axis=1)
cross_entropy_loss = tf.reduce_mean(-tf.log(likelihood))

minimize_op = tf.train.AdamOptimizer(1e-3).minimize(cross_entropy_loss)
global_init_op = tf.global_variables_initializer()
global_init_op2 = tf.variables_initializer(tf.global_variables())

#compute train accuracy
batch_size = 100
pred_class = tf.argmax(predict, axis=1)
true_class = tf.argmax(y_pl, axis =1 )
tf_correct = tf.reduce_sum(tf.cast(tf.equal(pred_class,true_class),tf.int32))
accuracy = tf_correct/batch_size
```

```python
# 2. Run the graph



nb_train_batch = int(x_train.shape[0]/batch_size) # the number of train batches
nb_remainder = x_train.shape[0]%batch_size

with tf.Session() as sess:
    sess.run(global_init_op)
    nb_epoch = 1
    for _ in range(nb_epoch):
        # permutated index
        perm_idx = np.random.permutation(x_train.shape[0]) #permutated index
        x_train = x_train[perm_idx]
        y_train_oh = y_train_oh[perm_idx]
        y_train = y_train[perm_idx]
        for i in range(nb_train_batch):
            x_batch = x_train[i*batch_size:(i+1)*batch_size]
            y_batch = y_train_oh[i*batch_size:(i+1)*batch_size]
            
            feed_dict = {x_pl : x_batch , y_pl : y_batch}
            fetch = [minimize_op, cross_entropy_loss, accuracy]
            _, train_loss, train_accuracy= sess.run(fetch, feed_dict)
            
            if (i+1)%10==0:
                print('step : %d, train loss : %.4f, train_accuracy : %.4f '%((i+1)/10,train_loss,train_accuracy))
                
    #compute test accuracy
    nb_test_batch = int(x_test.shape[0]/batch_size) # the number of test batches
    nb_test_remainder = x_test.shape[0]%batch_size
    nb_correct =0
    for i in range(nb_test_batch):
        x_batch = x_test[i*batch_size:(i+1)*batch_size]
        y_batch = y_test_oh[i*batch_size:(i+1)*batch_size]
        
        feed_dict = {x_pl : x_batch, y_pl : y_batch}
        nb_correct += sess.run(accuracy , feed_dict)*batch_size
    if nb_test_remainder !=0:
        x_batch = x_test[nb_test_batch*batch_size: ]
        y_batch = y_test_oh[nb_test_batch*batch_size]
        feed_dict = {x_pl : x_batch, y_pl : y_batch}
        nb_correct += sess.run(accuracy , feed_dict)*nb_test_remainder
    test_accuracy = nb_correct/x_test.shape[0]
    print('Test accuracy : %.4f' % (test_accuracy))
    #print(output.shape)
    #print(type(output))
```
