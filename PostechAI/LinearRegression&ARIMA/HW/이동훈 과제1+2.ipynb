## 3. 다음을 구하세요.

### 3-1. 5개의 샘플을 이용한 가설 검정을 시행하세요.

    2-1-1. N(0.1,1)을 따르는 정규분포에서 5개의 샘플을 구하세요.

```python
n = 5
x=stats.norm.rvs(0.1,1,n)
```

    2-1-2. 2-1-1에서 구한 데이터를 이용하여 유의도 0.05를 가지는 다음의 가설 검정을 시행하세요.
   1. $H_0:$ 평균 = 0
   2. $H_1:$ 평균 > 0

```python
alpha = 0.05
mu = np.mean(x)
Z = np.sqrt(n)*(mu-0)
p_val = 1 - stats.norm.cdf(Z)
if p_val < alpha :
    print("reject", format(p_val))
else:
    print("accept", format(p_val))
```
