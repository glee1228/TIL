### 4.2.1 나이 이상치 파악

### 4.2.2 나이 다시 계산 (만 나이 계산)

```python
# 참고: 오늘 날짜 구하는 방법
dt.date.today()
```

```python
def f(x):
    if x.month > dt.date.today().month: # 아직 생일 안 지남
        return dt.date.today().year - x.year - 1
    elif x.month < dt.date.today().month: # 생일 지남
        return dt.date.today().year - x.year
    else:
        if x.day <= dt.date.today().day:
            return dt.date.today().year - x.year
        else:
            return dt.date.today().year - x.year - 1
        
df['Age'] = df['Birthdate'].apply(f)
```

```python
df.sort_values('Age')
```

# 6. Data cleansing (Income)
데이터를 입력하는 사람에 따라 단위를 다르게 생각하고 값을 입력할 수 있다. 예를 들어, 소득이 20000달러일 때, 어떤 사람은 20000이라고 입력하고 어떤 사람은 1000달러를 기본 단위로 생각하여 20를 입력할 수 있다. 이를 고려하여 이상치를 찾고, 이상치라고 판단할 수 있는 근거를 제시하라. 또한, 이 이상치를 어떻게 처리할지 제시하고 처리하라.
1. 이상치라고 판단되는 값 확인
2. 이상치라고 판단할 근거 제시 (일반적인 값이라고 판단되는 데이터를 찾고, 비교)
3. 이상치 처리

```python
df.sort_values('Income')
```

Income으로 정렬된 결과를 봤을 때, 1000이하로 입력된 값들을 확인할 수 있음. 이 값들이 기본 단위를 1000 달러로 생각하고 잘못 입력한 값들로 의심됨

```python
df[df.Income < 1000].describe()
```

```python
df[df.Income > 1000].describe()
```

1000달러를 기준으로 이상치로 의심되는 데이터와 정상으로 생각되는 데이터를 나눠서 데이터의 descriptive statistics를 비교해 보면, <br>
우선, Income이 1000달러 미만인 경우는 10건으로 매우 적은 양이며,<br>
이상치로 의심되는 데이터들의 총 구매액(AmountSpent)이 평균적으로 더 많다. AmountSpent 값이 더 큰데 Income 값이 훨씬 작다는 것은 문제가 있다고 판단할 수 있다.

이상치 처리는 
1. 1000달러 미만의 데이터를 제거하거나
2. 1000달러 미만 데이터의 경우 단위 오류로 판단하고 1000을 곱해서 데이터를 보정하는 두 가지 방법으로 할 수 있다.

```python
#df = df[df.Income > 1000]
```

```python
df.loc[df.Income < 1000, 'Income'] = df.loc[df.Income < 1000, 'Income']*1000
```
