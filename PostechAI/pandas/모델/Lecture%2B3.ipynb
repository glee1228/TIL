# 회귀 분석(Regression Analysis)

회귀 분석은 종속 변수 Y를 독립 변수 X로 설명하는 자료 분석 방법이다.
잘 알려진 회귀 분석 기법으로는 **선형 회귀(LInear Regression)**분석, **로지스틱 회귀(Logistic Regression)**분석, **가우시안 프로세스 회귀(Gaussian Process Regression)**분석 등이 있다. 잘 알려진 회귀 분석 기법들은, Python Library에 잘 구현되어 있다.

회귀 분석에 필요한 모듈 가져오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from statsmodels import regression
from sklearn.datasets import load_boston
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel as C
from scipy import stats
import statsmodels.api as sm
```

## 선형 회귀 분석
선형 회귀 분석 에서 추세선 자체는 사람의 손으로 쉽게 구할 수 있지만, 이와 관련된 기타 통계값들은 손으로 구하기 쉽지않다. Python모듈에서는 관련된 주요 통계값들도 한번에 알려준다.

```python
def linreg(X,Y):
    # Running the linear regression
    X1 = sm.add_constant(X)
    model = regression.linear_model.OLS(Y, X1).fit()
    a = model.params[0]
    b = model.params[1]

    # Return summary of the regression and plot results
    X2 = np.linspace(X.min(), X.max(), 100)
    Y_hat = X2 * b + a
    plt.scatter(X, Y, alpha=0.3) # Plot the raw data
    plt.plot(X2, Y_hat, 'r', alpha=0.9);  # Add the regression line, colored in red
    #plt.xlabel('X Value')
    #plt.ylabel('Y Value')
    plt.show()
    return model.summary()
```

```python
X = np.random.rand(100)
Y = np.random.rand(100)
linreg(X, Y)
```

```python
Y = X + 0.2*np.random.randn(100)

linreg(X,Y)
```

```python
boston = load_boston()
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
data["price"] = boston.target
print(data.head())
```

```python
X = data["RM"]
Y = data["price"]
linreg(X,Y)
```

## 로지스틱 회귀 분석
로지스틱 회귀 분석은 종속 변수가 범주형 데이터인 경우, 특히 두가지 값만 가지는 경우(0,1)에 사용되는 **분류(Classification)**알고리즘중 하나이다. $Y=1$일 확률을 p라고 했을 때, p의 로짓 변환, $\ln \frac{p}{1-p}$를 X에 대하여 선형 함수로 근사하는 기법이다. 선형 회귀와 유사한 것 같지만, 선형 회귀와는 다르게, 사람 손으로 푸는 것은 거의 불가능 하다. 주로 최대 가능도 추정 기법을 이용하여 선형 함수로 근사하며, 선형 회귀와 마찬가지로 Python에서 구현이 잘 되어 있다.

### 로지스틱 함수

```python
x = np.linspace(-6, 6, 10000)
logit = lambda x: (1 + np.exp(-x))**(-1)
y = logit(x)
plt.plot(x, y, 'b', linewidth = 3, alpha=0.9)
plt.show()# Add the regression line, colored in red
```

### 선형 회귀와 로지스틱 회귀의 비교

```python
xmin, xmax = -5, 5
n_samples = 100
np.random.seed(0)
X = np.random.normal(size=n_samples)
y = (X > 0).astype(np.float)
X[X > 0] *= 4
X += .3 * np.random.normal(size=n_samples)

X = X[:, np.newaxis]
# run the classifier
clf = linear_model.LogisticRegression(C=1e5)
clf.fit(X, y)

# and plot the result
plt.scatter(X.ravel(), y, color='black', zorder=20)
X_test = np.linspace(-5, 10, 300)
```

```python
def model(x):
    return 1 / (1 + np.exp(-x))
loss = model(X_test * clf.coef_ + clf.intercept_).ravel()
plt.plot(X_test, loss, color='red', linewidth=3)

crit = -clf.intercept_ / clf.coef_
ols = linear_model.LinearRegression()
ols.fit(X, y)
plt.plot(X_test, ols.coef_ * X_test + ols.intercept_, linewidth=3)
plt.axhline(.5, color='.5')
```

```python
#plt.ylabel('y')
#plt.xlabel('X')
plt.xticks(range(-5, 10))
plt.yticks([0, 0.5, 1])
plt.ylim(-.25, 1.25)
plt.xlim(-4, 10)
plt.vlines(crit,-.25,1.25, color='b', linestyle = '--')
plt.legend(('Logistic Regression Model', 'Linear Regression Model'),
           loc="lower right")
plt.show()
```

## 가우시안 프로세스 회귀
선형 회귀나 로지스틱 회귀 처럼 매개변수를 추정하는 기법이 아닌 함수 그자체를 추정하는 기법이다. 관측된 데이터에 noise가 없다고 가정하는 경우와, noise가 있다고 가정했을 때 두가지 경우가 있으며, Python에서 구현이 잘 되어 있다.

```python
np.random.seed(1)
def f(x):
    """The function to predict."""
    return x * np.sin(x)
# ----------------------------------------------------------------------
#  First the noiseless case
X = np.atleast_2d([1., 3., 5., 6., 7., 8.]).T

# Observations
y = f(X).ravel()

# Mesh the input space for evaluations of the real function, the prediction and
# its MSE
x = np.atleast_2d(np.linspace(0, 10, 1000)).T

# Instanciate a Gaussian Process model
kernel = C(1.0, (1e-3, 1e3)) * RBF(10, (1e-2, 1e2))
gp = GaussianProcessRegressor(kernel=kernel, n_restarts_optimizer=9)

# Fit to data using Maximum Likelihood Estimation of the parameters
gp.fit(X, y)

# Make the prediction on the meshed x-axis (ask for MSE as well)
y_pred, sigma = gp.predict(x, return_std=True)

# Plot the function, the prediction and the 95% confidence interval based on
# the MSE

plt.plot(x, f(x), 'r:', label=u'$f(x) = x\,\sin(x)$')
plt.plot(X, y, 'r.', markersize=10, label=u'Observations')
plt.plot(x, y_pred, 'b-', label=u'Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.ylim(-10, 20)
plt.legend(loc='upper left')
plt.show()
```

```python
np.random.seed(1)

X = np.linspace(0.1, 9.9, 20)
X = np.atleast_2d(X).T

# Observations and noise
y = f(X).ravel()
dy = 0.5 + 1.0 * np.random.random(y.shape)
noise = np.random.normal(0, dy)
y += noise

# Instanciate a Gaussian Process model
gp = GaussianProcessRegressor(kernel=kernel, alpha=(dy / y) ** 2,
                              n_restarts_optimizer=10)

# Fit to data using Maximum Likelihood Estimation of the parameters
gp.fit(X, y)

# Make the prediction on the meshed x-axis (ask for MSE as well)
y_pred, sigma = gp.predict(x, return_std=True)

# Plot the function, the prediction and the 95% confidence interval based on
# the MSE

plt.plot(x, f(x), 'r:', label=u'$f(x) = x\,\sin(x)$')
plt.errorbar(X.ravel(), y, dy, fmt='r.', markersize=10, label=u'Observations')
plt.plot(x, y_pred, 'b-', label=u'Prediction')
plt.fill(np.concatenate([x, x[::-1]]),
         np.concatenate([y_pred - 1.9600 * sigma,
                        (y_pred + 1.9600 * sigma)[::-1]]),
         alpha=.5, fc='b', ec='None', label='95% confidence interval')
plt.ylim(-10, 20)
plt.legend(loc='upper left')

plt.show()
```

# ARIMA 모형
ARIMA 모형은 시계열 분석에 사용되는 대표적인 확률 모델이다. 시계열 모델의 규칙성을 설명하는 **AR(p)**모델, 불규칙한 패턴을 설명하는 **MA(q)** 모델이 있으며 **ARMA(p,q)** 모델은 AR(p)와 MA(q)모델의 합성이다. 마지막으로 **SARIMA(p,d,q)**모델은 ARMA 모델을 좀 더 일반화 한 모델이다.

주어진 시계열 데이터가 있을 때, 해당 데이터가 ARIMA모델로 설명이 된다고 가정하고, ARIMA모델에 해당하는 계수들을 추정하는 기법이 널리 알려져 있다. ARIMA 모델의 매개 변수인 p,d,q는 사람이 직접 정해줘야 하는 값이며, 대부분의 정제된 데이터는 (2,2,2)선에서 설명이 된다.

## AR 모형
현재의 시계열 값을 이전 시점 p개의 데이터로 선형 근사하는 모델이다. 즉,
$$X_t = a_0 + a_1 X_{t-1} + ... + a_p X_{t-p} + e_t$$
로 나타낸다.
선형 회귀와 크게 다를 것이 없으므로, $a_0$, ..., $a_p$를 구하는 것은 그리 어렵지 않지만, AR(p)가 의미있는 모델이 되기 위해서는 다음의 **정상성** 조건이 필요하다.
#### 정상성 조건
다항식 $-1 + a_1 x + a_2 x^2 + ... + a_p x^p = 0$의 해의 절대값이 모두 1보다 커야한다.

시계열 데이터를 다루기 위해서는 다음의 모듈을 추가로 불러와야 한다.

```python
import statsmodels.api as sm
import statsmodels.tsa as tsa
```

### AR 모델 생성

```python
np.random.seed(100)

# This function simluates an AR process, generating a new value based on historial values,
# autoregressive coefficients b1 ... bk, and some randomness.
def AR(b, X, mu, sigma):
    l = min(len(b)-1, len(X))
    b0 = b[0]
    
    return b0 + np.dot(b[1:l+1], X[-l:]) + np.random.normal(mu, sigma)

b = np.array([0, 0.8, 0.1, 0.05])
p=len(b)-1
X = np.array([1])
mu = 0
sigma = 1

for i in range(10000):
    X = np.append(X, AR(b, X, mu, sigma))

plt.figure(figsize=(15,5))
plt.plot(X[1000:1300], 'blue', alpha = 0.8)
plt.xlabel('Time')
plt.ylabel('AR({})'.format(p))
plt.show()
```

## MA 모형
현재의 시계열 값의 에러를 이전 시점 p개의 에러로 선형 근사하는 모델이다. 즉,
$$X_t = b_1 e_{t-1} + ... + b_q e_{t-q} + e_t$$
로 나타낸다.
유한한 개수의 확률 변수의 선형 결합 이므로, 각각의 시점에서의 분산이 유한하기 때문에, AR 모델과는 다르게 정상성 조건이 필요하지 않다. 다만 특정 시점의 에러가 얼마인지를 데이터로 부터 알 수 없기 때문에 손으로 직접 계수를 추정하기 불가능하다.

### MA 모형 생성

```python
np.random.seed(100)
def AR(b, X, mu, sigma):
    l = min(len(b)-1, len(X))
    b0 = b[0]
    
    return b0 + np.dot(b[1:l+1], X[-l:]) + np.random.normal(mu, sigma)

def MA(theta, error, mu=0, sigma=1):
    l = min(len(theta)+1 , len(error)) - 1 
    return np.dot(theta[0:l], error[-l:])


theta = np.array([1,1,1])
q=len(theta)
error = np.random.normal(0,1)
cal_error = error

mu = 0
sigma = 1

for i in range(10000):
    error = np.append(error, np.random.normal(mu, sigma))
    cal_error = np.append(cal_error, MA(theta, error, mu, sigma))

plt.plot(cal_error[1000:1300])
plt.xlabel('Time')
plt.ylabel('MA({})'.format(q));
plt.show()
```

## ARMA 모형
AR모델과 MA모델의 결합이며, 다음과 같이 나타낸다.
$$X_t = a_0 + a_1 X_{t-1} + ... + a_p X_{t-p} + b_1 e_{t-1} + ... + b_q e_{t-q} + e_t$$
b_1 e_{t-1} + ... + b_q e_{t-q} +b_1 e_{t-1} + ... + b_q e_{t-q} +
AR모델을 포함하기 때문에, **정상성** 조건이 필요하다.

### ARMA 모형 생성

```python
np.random.seed(100)
def ARMA(b, X, cal_error):
    l = min(len(b)-1, len(X))
    b0 = b[0] + cal_error
    
    return b0 + np.dot(b[1:l+1], X[-l:])

X = np.array([1])
b = np.array([0, 0.8, 0.1, 0.05])
p=len(b)-1
theta = np.array([1,1,1])
q=len(theta)-1
error = np.random.normal(0,1)
cal_error = error

mu = 0
sigma = 1

for i in range(10000):
    error = np.append(error, np.random.normal(mu, sigma))
    cal_error = MA(theta, error, mu, sigma)
    X = np.append(X, ARMA(b, X, cal_error))

plt.figure(figsize=(15,5))
plt.plot(X[1000:1300])
plt.xlabel('Time')
plt.ylabel('ARMA({}, {})'.format(p,q));
plt.show()
```

## ARIMA, SARIMA 모형
시계열의 정상성은 깨졌지만 시계열을 d번 차분한 데이터가 ARMA 모델을 따르는 모델이다. 일반적인 시계열 데이터의 경우 정상성이 깨져 있어도 차분($X'_t := X_t - X_{t-1}$ 하면 정상성이 보이는 경우가 많기 때문에 자주 사용된다. 차분 한뒤 원래 데이터로 환원 할때는 누적 합계 $X_t = \sum_{s=1}^t X'_s + X_0$를 이용한다.

SARIMA모형은 ARIMA 모델에 주기적인 요소(Seasonal effect)를 고려한 모델로서, 온도와 같이 일정한 주기가 보이는 데이터의 예측에 사용되는 모델이다.

### ARIMA 모델의 생성

```python
np.random.seed(100)
X = np.array([1])
b = np.array([0, 0.8, 0.1, 0.05])
p=len(b)-1
theta = np.array([1,0.3,0.3])
q=len(theta)-1
error = np.random.normal(0,1)
cal_error = error

mu = 0
sigma = 1

for i in range(10000):
    error = np.append(error, np.random.normal(mu, sigma))
    cal_error = MA(theta, error, mu, sigma)
    X = np.append(X, ARMA(b, X, cal_error))

X=np.cumsum(X)

plt.figure(figsize=(15,5))
plt.plot(X[1000:1300])
#plt.xlabel('Time')
#plt.ylabel('ARMA({},1, {})'.format(p,q));
plt.show()
```

## ARIMA 모델의 적용
월별 대기의 $CO_2$ 농도량 데이터를 이용하여, 다음 달의 대기의 $CO_2$ 농도량을 예측해보자.

### Pandas에 내장된 데이터 불러오기

```python
data = sm.datasets.co2.load_pandas()
y = data.data
# The 'MS' string groups the data in buckets by start of the month
y = y['co2'].resample('MS').mean()

# The term bfill means that we use the value before filling in missing values
y = y.fillna(y.bfill())
y.plot()
plt.xlabel('Date')
plt.ylabel('CO2 levels');
plt.show()
```

### p,d,q 정하기
p,d,q,를 정하는 방법중의 하나는 **AIC**(Akaike Information Criterioin)을 주로 사용한다. 각각의 모델에 대한 AIC값을 계산한 후 값이 가장 낮은 모델을 선택한다. 

#### 매개변수들의 집합 생성
$CO_2$ 농도가 1년의 주기를 띄고 상승 하강하는 것이 보이므로, Seasonal effect를 12로 준다.

```python
import itertools
p = d = q = range(0, 2)

# Generate all different combinations of p, q and q triplets
pdq = list(itertools.product(p, d, q))

# Generate all different combinations of seasonal p, q and q triplets
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

#### 각각의 매개변수에 대한 AIC값 계산

```python
import warnings

warnings.filterwarnings("ignore") # specify to ignore warning messages

AIC = pd.DataFrame({"param":[], "param_seasonal":[], "aic":[]})
for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(y,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            TEMP = pd.DataFrame({"param":[param], "param_seasonal":[param_seasonal], "aic":[results.aic]})
            AIC  = pd.concat([AIC, TEMP]).reset_index(drop=True)
        except:
            continue
AIC.sort_values("aic").head(5)
```

(1,1,1)$\times$(1,1,1,12)일때 AIC값이 가장 낮게 나왔으므로, 해당 매개 변수를 적용한 뒤 $CO_2$농도를 예측 한다.

```python
mod = sm.tsa.statespace.SARIMAX(y,
                                order=(1, 1, 1),
                                seasonal_order=(1, 1, 1, 12),
                                enforce_stationarity=False,
                                enforce_invertibility=False)

results = mod.fit()

pred = results.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=False)
pred_ci = pred.conf_int()
```

```python
ax = y['1990':].plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7)

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend(loc="lower right")


plt.show()
```
