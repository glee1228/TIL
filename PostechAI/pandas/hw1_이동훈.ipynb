# 4. Data cleansing - Birthdate & Age (16점)

## 4.1 Birthdate (11점)

### 4.1.1 Birthdate 타입 변경 (4점)

```python
# 날짜 데이터를 다루기 위한 모듈
import datetime as dt
```

```python
data['Birthdate'].tail()
```

```python
data['Birthdate']=pd.to_datetime((data['Birthdate']))
data.head()
```

```python
def f(x):
    if x.year>2018:
        return dt.datetime(x.year-100,x.month,x.day)
    else :
        return x
```

```python
data.Birthdate=data.Birthdate.apply(f)
```

```python
data.head(20)
```

Birthdate의 형태가 'M.D.Y'에서 'Y-M-D'로 바뀐 것을 확인할 수 있음. 데이터 타입 또한 datetime으로 변경됨.<br>
하지만 첫 줄에서 볼 수 있듯이, 2044-10-26과 같은 불가능한 생일의 형태가 존재함. 이런 데이터는 1944-10-26으로 바꿔줘야 함.

# 6. Data cleansing - Income (7점)
데이터를 입력하는 사람에 따라 단위를 다르게 생각하고 값을 입력할 수 있다. 예를 들어, 소득이 20000달러일 때, 어떤 사람은 20000이라고 입력하고 어떤 사람은 1000달러를 기본 단위로 생각하여 20를 입력할 수 있다. 이를 고려하여 이상치를 찾고, 이상치라고 판단할 수 있는 근거를 제시하라. 또한, 이 이상치를 어떻게 처리할지 제시하고 처리하라.
1. 이상치라고 판단되는 값 확인
2. 이상치라고 판단할 근거 제시 (일반적인 값이라고 판단되는 데이터를 찾고, 비교)
3. 이상치 처리

```python
data.sort_values('Income',ascending=True).head(20)
```

```python
#평균이 90418이고 하위 25% 값이 56175인걸로 미루어 보아 최소 만 단위의 데이터가 존재해야함을 알수있다
data['Income'].describe()
#10000미만의 세자리 데이터에 1000을 곱하여 데이터를 정제한다.
data.Income=data.Income.apply(lambda x :x*1000 if x<10000 else x)
data.Income.value_counts()
```
