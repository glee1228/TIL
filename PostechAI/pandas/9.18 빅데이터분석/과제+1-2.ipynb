    2. N(1,1)

```python
mu=1
sigma = 1
f = lambda x: 1/np.sqrt(2*np.pi*sigma)*np.exp(-0.5*(x-mu)**2/sigma)

x = np.linspace(-10,10, 10000) #-5에서 5까지 균등한 한격의 10000개의 수를 만든다.
plt.plot(x,f(x), linewidth = 5) #linewidth는 그래프에 그리는 라인의 두께를 의미한다.
plt.show()
```

    3. 1/2의 확률로 N(1,1)과 N(3,2^2)를 따르는 분포 

```python
mu1 = 1
sigma1 = 1
mu2 = 3
sigma2 = 2**2
f = lambda x : 1/np.sqrt(2*np.pi*sigma1)*np.exp(-0.5*(x-mu1)**2/sigma1)
f2 = lambda x : 1/np.sqrt(2*np.pi*sigma2)*np.exp(-0.5*(x-mu2)**2/sigma2)

x = np.linspace(-10,10,10000)
plt.plot(x,f(x),linewidth=5)
plt.plot(x,f2(x),linewidth=5)
plt.show()
```
